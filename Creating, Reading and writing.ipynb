# Manipulating and Indexing

```python
# Set_index()Methods

reviews.set_index("title")
```

```python
# Conditional Selection
# Lets check if wine in Italy

reviews.country == 'Italy' # this produce a series of True/false Booleans
```

```python
#using loc to get relevant data

reviews.loc[reviews.country == 'Italy']
```

```python
# Select review with great or equal to 90 points scale

reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
# To find wine rated above average with use pipe |

reviews.loc[(reviews.country == 'Italy') |(reviews.points >= 90)]
```

```python
# Example of isin selector, will tell if value is in list of value

reviews.loc[reviews.country.isin(['Italy','France'])]
```

```python
# Using notnull

reviews.loc[reviews.price.notnull()]
```

# Maps

```python
# using mapping function

review_points_mean = reviews.points.mean()
reviews.points.map(lambda p:p - review_points_mean)
```

```python
# Apply is equivalent method if we want to transform a whole dataframe by a customer method on each row

def remean_points(row):
    row.points = row.points - reviews_points_mean
    return row

reviews.apply(remean_points, axis = 'columns')
```

```python
reviews.head(2)
```

```python
# quicker way to remeaning point col using pandas


reviews_points_mean = reviews.points.mean()
reviews.points - review_points_mean
```

# Grouping

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
# select name of first wine review from dataset

reviews.groupby('winery').apply(lambda df:df.title.iloc[0])
```

```python
# more granular way # group more than one column

reviews.groupby(['country','province']).apply(lambda df:df.loc[df.points.idxmax()])


```

```python
# another groupby methon call agg()

reviews.groupby(['country']).price.agg([len,min,max])
```

# Multi-indexes

```python
# multi index has multiple levels

countries_reviewed = reviews.groupby(['country','province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
#converting back to regular index using reset_index()

countries_reviewed.reset_index()

```

# Sorting

```python
# using sort_index()

countries_reviewed.sort_index()
```
